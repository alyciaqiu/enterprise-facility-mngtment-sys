```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# read all floor clean data
data = pd.read_csv('ALL_Floor_clean_data_new.csv')
data
```

```python
# read macid data
macid = pd.read_csv('device_list.csv')
macid
```

```python
macid['Type'].value_counts()
```

```python
# find mac address under Fixed category
macid = macid[macid['Type'] == 'Fixed']
macid
```

```python
# all the records under Fixed category
data = data[data['ClientMacAddr'].isin(macid['ClientMacAddr'])]
data
```

```python
# check
data.groupby('ClientMacAddr').nunique()
```

```python
# read count of day
daily_data = pd.read_csv('ALL_Floor_data_count_of_day.csv')
daily_data
```

```python
# delete the macid only appeared one day
del_data =  daily_data[daily_data['count_of_day']==1]['ClientMacAddr']
data = data[~data['ClientMacAddr'].isin(del_data)]
data
```

```python
# count
mac_data = data.groupby('ClientMacAddr').nunique().reset_index()
mac_data
```

```python
mac_data.to_csv('Fixed_checked_macid.csv', index=False)
```
