```python
import pandas as pd
import numpy as np
```

```python
employee_identity = pd.read_csv('employee_identity.csv')
macid = pd.read_csv('final_id_3.csv')
equip_status = pd.read_csv('equip_status.csv')
equip_info = pd.read_csv('equip_info.csv')
employee_action = pd.read_csv('employee_action.csv')
```

```python
equip_info
```

```python
equip_status['Priority'] = equip_status['Prority']
equip_status = equip_status.drop('Prority', axis=1)
equip_status
```

```python
equip_status = equip_status[~equip_status['Priority'].isna()]
equip_status
```

```python
equip = pd.merge(equip_status, equip_info, on=['EquipmentID','Process'], how='left')
equip = equip[['EquipmentID', 'EquipmentState', 'localtime', 'Priority', 'Process', 'Level', 'ZoneID']]
equip
```

```python
macid['EquipmentID'] = macid['ClientMacAddr'].apply(lambda x: 'P001' if x=='88:64:a6:49:88:11' else 'P002' if x=='80:58:f8:5a:0d:88' else 'P004')
macid = macid.drop('Unnamed: 0', axis=1)
macid
```

```python
employee = pd.merge(macid, employee_identity, on='ClientMacAddr', how='left')
employee
```

```python
equip_employee = pd.merge(equip, employee, on='EquipmentID', how='left')
equip_employee
```

```python
equip_employee_action = pd.merge(equip_employee, employee_action, on=['EmployeeRole', 'Process', 'EquipmentState', 'Priority'], how='left')
equip_employee_action
```

```python
def employee_email_content (equip_employee_action):
    for i, row in equip_employee_action.iterrows():
        print('Send to:' + row['Email'])
        employee = f"Dear {row['EmployeeRole']} {row['EmployeeName']}, \n"
        print(employee)
        alert = f"The state of equipment ({row['EquipmentID']}) changes to {row['EquipmentState']} at {row['localtime']}."
        print(alert)
        print('The priority is ' + row['Priority'] + '.')
        location = f"The location is {row['Level']}, and the zone id is {row['ZoneID']}."
        print(location)
        action = f"What you need to do is the following step(s): \n{row['Action']} \n"
        print(action)
        print('Please investigate on-site as soon as possible! \n\n')
```

```python
employee_email_content(equip_employee_action)
```
