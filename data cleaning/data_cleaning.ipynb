```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = pd.read_csv('ALL_Floor_data.csv')
data
```

```python
data = data.drop(labels='Unnamed: 0', axis=1)
data
```

```python
# data cleaning
# delete data containing NA
data = data.dropna()
data
```

```python
# delete duplicate rows
data = data.drop_duplicates()
data
```

```python
# delete clientMacAddr = 00:00:00:00
data = data[data['ClientMacAddr'] != '00:00:00:00:00:00']
data
```

```python
# delete outliers out of boundary
df_1 = data

# For the first rectangle, I filter out  
# Logic: left min, right max
df_2 = df_1.query("lat>=51.46051343244503 and lat<=51.46094763183397")
df_2 = df_2.query("lng>=-0.9326075994237232 and lng<=-0.9322332468412318")
df_2.reset_index(drop=True)

# For the second rectangle, I filter out
df_3 = df_1.query("lat>=51.460257062787136 and lat<=51.46055317954987")
df_3 = df_3.query("lng>=-0.9333046313023279 and lng<=-0.9322952552121134")
df_3.reset_index(drop=True)

#Merging df_2 and df_3
df_1 = pd.concat([df_2, df_3], ignore_index=True).drop_duplicates()

data = df_1.reset_index(drop=True)
```

```python
data
```

```python
# time accurate to seconds
# each mac address only keeps a unique row of data per second
data['localtime'] = data['localtime'].apply(lambda x:x[:19])
data
```

```python
data['localtime'] = pd.to_datetime(data['localtime'])
data
```

```python
data = data.sort_values('localtime',ascending = True).groupby(['ClientMacAddr','localtime']).first().reset_index()
data
```

```python
# delete mac address with only one record
del_data = data.groupby('ClientMacAddr').count().reset_index()
del_data = del_data[del_data['localtime']==1]['ClientMacAddr']
data = data[~data['ClientMacAddr'].isin(del_data)]
data
```

```python
# save as csv
data.to_csv('ALL_Floor_clean_data_new.csv', index=False)
```
